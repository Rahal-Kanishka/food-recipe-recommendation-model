```python
data = pd.read_excel('RAW_recipes_original.xlsx', sheet_name = 0, index_col=1);
```

## filter out data

```python
filtered = data.copy()[data['n_ingredients'] < 4] 
```

```python
filtered.shape
```

## process column values

```python
filtered["tags"] = filtered["tags"].str.replace('[', '').str.replace(']', '').str.replace("'", "")
```

```python
filtered["nutrition"] = filtered["nutrition"].str.replace('[', '').str.replace(']', '').str.replace("'", "")
```

```python
filtered["ingredients"] = filtered["ingredients"].str.replace('[', '').str.replace(']', '').str.replace("'", "")
```

```python
filtered["steps"] = filtered["steps"].str.replace('[', '').str.replace(']', '').str.replace("'", "")
```

```python
filtered.head()
```

```python
def process_csv_data(column_name: str, data_frame: any):
    new = data_frame[column_name].str.split(pat = ",", expand=True).apply(lambda x : x.value_counts(), axis = 1).fillna(0).astype(int)
    final = pd.concat([data_frame, new], axis = 1)
    return final
```

## process column data

```python
filtered = process_csv_data('tags',filtered)
```

```python
# filtered = process_csv_data('ingredients',filtered)
```

```python
filtered.head()
```

```python
filtered.info()
```

```python
filtered.head(10)
```

## process datetime value

```python
filtered["time_dif"] = pd.Timestamp.now().normalize() - filtered["submitted"]
```

```python
filtered["time_dif"]
```

## Normalize the data using Max value

```python
filtered["scaled_minutes"] = filtered["minutes"] / filtered["minutes"].abs().max()
```

```python
filtered["scaled_time_dif"] = filtered["time_dif"] / filtered["time_dif"].abs().max()
```

```python
filtered["scaled_n_steps"] = filtered["n_steps"] / filtered["n_steps"].abs().max()
```

```python
filtered["scaled_n_ingredients"] = filtered["n_ingredients"] / filtered["n_ingredients"].abs().max()
```

```python
filtered.sort_values('id')
```

## drop unwanted columns

```python
final = filtered.drop(columns=['name','tags', 'nutrition', 'ingredients', 'minutes', 'submitted','time_dif', 'n_ingredients', 'n_steps', 'steps', 'description', 'contributor_id'])
```

```python
final.shape
```

```python
final.head(10)
```

## Perform Fit transform

```python
from sklearn.metrics.pairwise import cosine_similarity
```

## Create a matrix

```python
matrix = final.to_numpy()
```

```python
matrix
```

```python
matrix.shape
```

```python
similarities = cosine_similarity(matrix)
```

```python
similarities
```

```python
similarities.shape
```

## Save to Parquet File

```python
result = pd.DataFrame(similarities) #remove the .T here
# result.to_csv('output/similarities-second.csv')
result.to_parquet("output/large.parquet", compression=None, index=True)
print('parquet data saved')
```

## Save filtered dataset

```python
result = pd.DataFrame(filtered) #remove the .T here
# result.to_csv('output/similarities-second.csv')
result.sort_values('id')
result['row_index'] = range(0, len(result))
# result.set_index("row_index", inplace=True)
result.to_parquet("output/filtered-dataset.parquet", compression=None, index=True)
print('filtered dataset saved')
```
